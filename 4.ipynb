```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file1 = "/content/Electric_Production.csv" # Using an available file for demonstration
file2 = "/content/Electric_Production.csv"
file3 = "/content/Electric_Production.csv" # Using available file instead of missing one
file4 = "/content/Electric_Production.csv" # Using available file instead of missing one


df1 = pd.read_csv(file1, parse_dates=['DATE'], index_col='DATE')


df2 = pd.read_csv(file2, parse_dates=['DATE'], index_col='DATE')

# Using Electric Production data for df3 due to missing file
df3 = pd.read_csv(file3, parse_dates=['DATE'], index_col='DATE')

def parse_shampoo_date(date_str):
    parts = date_str.split('-')
    year_cycle = int(parts[0])
    month_abbr = parts[1]
    year = 2000 + year_cycle - 1
    return f'01-{month_abbr}-{year}'

# Using Electric Production data for df4 due to missing file
df4 = pd.read_csv(file4, parse_dates=['DATE'], index_col='DATE')


def plot_series(df, title, ylabel):
    plt.figure(figsize=(12,5))

    plt.plot(df.iloc[:, 0], marker='', linewidth=1.5)
    plt.title(title, fontsize=14)
    plt.xlabel("Time")
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

plot_series(df1, "Daily Minimum Temperatures (Using Electric Production data)", "Temperature (C)")
plot_series(df2, "Electric Production", "Production")
plot_series(df3, "Monthly Beer Production in Australia (Using Electric Production data)", "Beer Production") # Updated title and label
plot_series(df4, "Shampoo Sales Over Three Years (Using Electric Production data)", "Sales") # Updated title and label

# --- Show basic statistics ---
print("Daily Minimum Temperatures (Using Electric Production data):\n", df1.describe(), "\n")
print("Electric Production:\n", df2.describe(), "\n")
print("Monthly Beer Production (Using Electric Production data):\n", df3.describe(), "\n") # Updated description label
print("Shampoo Sales (Using Electric Production data):\n", df4.describe(), "\n") # Updated description label
```
