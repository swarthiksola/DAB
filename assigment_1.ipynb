```python
import numpy as np
import pandas as pd

# === 1. Data Creation and Categorization ===

# Sample student grades
grades = np.array([67, 45, 88, 50, 39, 72, 54, 33, 90, 47])

# Define a function to classify grades using a dynamic threshold
def classify_grades(grades, threshold=50):
    return pd.Series(np.where(grades >= threshold, "Pass", "Fail"), dtype="category")

# Classify with default threshold (50)
categories = classify_grades(grades)

# === 2. Data Structure Analysis ===

print("Length of grades vector:", len(grades))
print("Data type of grades vector:", type(grades))
print("Data type of category vector:", type(categories))
print("Unique values in category vector:", categories.unique())

# === 3. Statistical Summary ===

mean_grade = np.mean(grades)
median_grade = np.median(grades)
pass_rate = (categories == "Pass").mean() * 100  # percentage
fail_count = (categories == "Fail").sum()

print("\n--- Statistical Summary ---")
print("Mean grade:", mean_grade)
print("Median grade:", median_grade)
print("Pass rate: {:.2f}%".format(pass_rate))
print("Fail count:", fail_count)

# === 4. Interpretation Task (Dynamic Threshold Example) ===

# Example: Using a new threshold (e.g., 60)
new_threshold = 60
new_categories = classify_grades(grades, threshold=new_threshold)

new_pass_rate = (new_categories == "Pass").mean() * 100
new_fail_count = (new_categories == "Fail").sum()

print(f"\n--- With New Threshold = {new_threshold} ---")
print("New Pass rate: {:.2f}%".format(new_pass_rate))
print("New Fail count:", new_fail_count)
```
